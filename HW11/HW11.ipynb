```python
import numpy as np
import pandas as pd
pd.options.display.max_columns = 999
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
data = pd.read_csv('example.csv')
data.tail()
```

```python
data.describe()
```

```python
for column in data.columns:
    print(column, "mean: ",data[column].mean(),",",column,"std: ", data[column].std(),"\n")
```

> ### We can see that the data set is already normalized

```python
class_count = pd.value_counts(data.Class)
class_count.plot(kind = 'bar')
plt.title("Class Counts")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()

print(class_count[0], "for class 0 / ", class_count[1]," for class 1")
```

> Class Imbalance

```python
normal_data = data[data['Class'] == 0]
target_data = data[data['Class'] == 1]
```

```python
columns_without_class = list(data.columns)[:28]
for column in columns_without_class:
    sns.distplot(normal_data[column])
    sns.distplot(target_data[column])
    plt.show()
```

> ### Some features show different distribution depeding on class.

```python
for column in columns_without_class:
    plt.figure(figsize=(8,4))
    plt.subplot(1,2,1)
    sns.boxplot(normal_data[column])
    plt.title("normal")
    plt.subplot(1,2,2)
    sns.boxplot(target_data[column])
    plt.title("target")
    plt.tight_layout()
    plt.show()
```

```python
for column in columns_without_class:
    plt.figure(figsize=(8,4))
    plt.scatter(normal_data.V1, normal_data[column], alpha = 0.05)
    plt.scatter(target_data.V1, target_data[column], alpha = 0.05)
    plt.show()
```

> ### Let's look into the correlations between featrues

```python
target_data.drop("Class", axis = 1).corr('pearson')
```
