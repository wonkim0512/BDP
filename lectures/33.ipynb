```python
help(HTMLParser)
```

```python
class MyHTMLParser(HTMLParser):
    def handle_starttag(self, tag, attrs):
        print("Encountered a start tag: ", tag)
        
    def handle_endtag(self, tag):
        print("Encountered an end tag: ", tag)
        
    def handle_data(self, data):
        print("Encountered some data: ", data)
        
parser = MyHTMLParser()
parser.feed("<html><head>  <meta charset='utf-8'/><title>Required Tags</title></head>\
            <body><p>This is <b>bold</b> text. <br></p></body></html>")
```

```python
while "\n" in content:
    content.remove('\n')
content
```

```python
new_content = []
for line in content:
    if line.lstrip()[-1] == "\n":
        new_content.append(line.lstrip()[:-1])

new_content
    
```
